# Computation of eigenvalues and eigenvectors of H

```python
values={"E1":0, "E2":4, "V11":3, "V22":-3, "V12":sp.Rational(1, 5), "V21":sp.Rational(1, 5)}
Hnew = H.subs(values)
display(Math(f'H = {sp.latex(Hnew)}'))
evals = Hnew.eigenvals()
evects = Hnew.eigenvects()
print(f'type(evals): {type(evals)}')

# display(evals)
# display(evects)

evals_values = list(evals.keys())
x1, x2 = evals_values[0], evals_values[1] 
display(Math(f'x_1 = {sp.latex(x1)}'))
display(Math(f'x_2 = {sp.latex(x2)}'))

sp.print_latex(x1)
# sp.print_latex(x2)
# sp.print_latex(Hnew)

print('\n' + 'eigenvectors:')
# display(evects)
display(Math(f'{sp.latex(evects)}'))

# https://stackoverflow.com/questions/43689076/how-to-find-the-eigenvalues-and-eigenvectors-of-a-matrix-with-sympy
v1, v2 = [list(tup[2][0]) for tup in evects]
print('\n')
display(Math(f'v_1 = {sp.latex(v1)}'))
sp.print_latex(v1)
print('\n\n')
display(Math(f'v_2 = {sp.latex(v2)}'))
sp.print_latex(v2)
```

```python
def normalizeRows(x):
    """
    Implement a function that normalizes each row of the matrix x (to have unit length).
    
    Argument:
    x -- A numpy matrix of shape (n, m)
    
    Returns:
    x -- The normalized (by row) numpy matrix. You are allowed to modify x.
    """
    
    # Compute x_norm as the norm 2 of x. Use np.linalg.norm(..., ord = 2, axis = ..., keepdims = True)
    x_norm = np.linalg.norm(x, axis=1, keepdims=True)
    print("x_norm.shape:", x_norm.shape, "\n")
    
    # Divide x by its norm.
    x = x / x_norm

    return x
```

```python
# np.linspace(0, 1, 11)
n = 101
lamda_verdier = np.linspace(0, 1, n)
v1_verdier = np.array([[sp.simplify(lamda*v1[0]).subs(lamda, l), l] for l in lamda_verdier] )
v2_verdier = np.array([[sp.simplify(lamda*v2[0]).subs(lamda, l), l] for l in lamda_verdier] )
# print(v1_verdier)
type(v1_verdier)
v1_verdier.shape
v1_verdier = np.array(v1_verdier).astype(np.float64)
v2_verdier = np.array(v2_verdier).astype(np.float64)
v1_verdier = normalizeRows(v1_verdier)
v2_verdier = normalizeRows(v2_verdier)
plt.plot(lamda_verdier, v1_verdier[:, 0], color='blue')
plt.plot(lamda_verdier, v1_verdier[:, 1], color='green')
plt.show()

plt.plot(lamda_verdier, v2_verdier[:, 0], color='blue')
plt.plot(lamda_verdier, v2_verdier[:, 1], color='green')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

lamb = lamda_verdier

component11 = v1_verdier[:, 0]
component12 = v1_verdier[:, 1]
component21 = v2_verdier[:, 0]
component22 = v2_verdier[:, 1]


# Plot on the first subplot (axs[0])
axs[0].plot(lamb, component11, color='blue',label="|0>")
axs[0].plot(lamb, component12, color="green",label="|1>")
axs[0].plot(lamb, component11**2 + component12**2, label="check")
axs[0].set_xlabel('$\lambda$')
axs[0].set_ylabel('$E_1$')
axs[0].set_ylim(-1, 1.1)
# axs[0].set_aspect('equal', adjustable='box')
# axs[0].axis('square')
axs[0].legend()

# Plot on the second subplot (axs[1])
axs[1].plot(lamb, component21, color='blue',label="|0>")
axs[1].plot(lamb, component22, color="green", label="|1>")
axs[1].plot(lamb, component21**2 + component22**2, label="check")
axs[1].set_xlabel('$\lambda$')
axs[1].set_ylabel('$E_2$')
axs[1].set_ylim(-1, 1.1)
# axs[1].set_aspect('equal', adjustable='box')
axs[1].legend()

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.25)

plt.suptitle('Eigenstates', fontsize=15)
plt.savefig('partB_eigenvectors_graf_3.png')
plt.show()
```
